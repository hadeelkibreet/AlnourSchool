# Part 11: Customer Segmentation

22. If your dataset includes customer information, use clustering techniques to segment customers based on their purchasing behavior. Visualize the segments using a scatter plot or a bar plot.

```python
#importing the required libraries
import pandas as pd
#viz Libraries
import matplotlib.pyplot as plt
plt.style.use('ggplot')

#warnings
import warnings
warnings.filterwarnings("ignore")

#datetime
import datetime as dt
```

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
quant_vars = ['QUANTITYORDERED','PRICEEACH','SALES','MSRP']
df[quant_vars].describe()
```

```python
df.groupby(['YEAR_ID'])['MONTH_ID'].nunique()
```

```python
# Calculate Recency, Frequency and Monetary value for each customer
snapshot_date = df['ORDERDATE'].max() + dt.timedelta(days=1) #latest date in the data set
df_RFM = df.groupby(['CUSTOMERNAME']).agg({
    'ORDERDATE': lambda x: (snapshot_date - x.max()).days,
    'ORDERNUMBER': 'count',
    'SALES':'sum'})

#Renaming the columns
df_RFM.rename(columns={'ORDERDATE': 'Recency',
                   'ORDERNUMBER': 'Frequency',
                   'SALES': 'MonetaryValue'}, inplace=True)

df_RFM.head()
```

Recency/Frequency/MonetaryValue : Level 4 > Level 3> Level 2 > Level 1

Lower the recency, higher the Recency level
Higher the number of orders, higher the Frequency level
Higher the monetary value, higher the MonetaryValue level

```python
#Dividing into segments

# Create a spend quartile with 4 groups - a range between 1 and 5
MonetaryValue_quartile = pd.qcut(df_RFM['MonetaryValue'], q=4, labels=range(1,5))
Recency_quartile = pd.qcut(df_RFM['Recency'], q=4, labels=list(range(4, 0, -1)))
Frequency_quartile = pd.qcut(df_RFM['Frequency'], q=4, labels=range(1,5))


# Assign the quartile values to the Spend_Quartile column in data
df_RFM['R'] = Recency_quartile
df_RFM['F'] = Frequency_quartile
df_RFM['M'] = MonetaryValue_quartile

df_RFM.head()
```

```python
# Calculate RFM_Score
df_RFM['RFM_Score'] = df_RFM[['R','F','M']].sum(axis=1)
df_RFM.head()
```

RFM Score > 10 - High Value Customer
RFM SCore < 10 and RFM Score >= 6 - Mid Value Customer
RFM Score < 6 - Low Value Customer

```python
# Define rfm_level function
def rfm_level(df):
    if bool(df['RFM_Score'] >= 10):
        return 'High Customer'
    elif bool((df['RFM_Score'] < 10) & (df['RFM_Score'] >= 6)):
        return 'Mid Customer'
    else:
        return 'Low Customer'

# Create a new variable RFM_Level
df_RFM['RFM_Level'] = df_RFM.apply(rfm_level, axis=1)

# Print the header with top 5 rows to the console
df_RFM.head()
```

```python
import matplotlib.pyplot as plt

# Count the number of customers in each RFM level
rfm_counts = df_RFM['RFM_Level'].value_counts()
plt.figure(figsize=(5, 3))


# Create a horizontal bar chart
plt.barh(rfm_counts.index, rfm_counts.values, color='blue')

# Set the labels and title
plt.xlabel('Number of Customers')
plt.ylabel('RFM Level')
plt.title('RFM Level Distribution')

# Display the chart
plt.show()
```

# Part 13: Interactive Visualizations

24. Create an interactive visualization using libraries like Plotly or Bokeh to allow users to explore the data by filtering or selecting specific countries or time periods.

```python
# Create a time periods column
df['TIME_PERIOD'] = df['MONTH_ID'].astype(str) + '-' + df['YEAR_ID'].astype(str)

# Get unique country values
unique_countries = df['COUNTRY'].unique()

# Create a categorical column for 'COUNTRY'
df['COUNTRY_CODE'] = pd.Categorical(df['COUNTRY']).codes

# Create the dropdown widgets
countries = widgets.Dropdown(options=unique_countries)
time_periods = widgets.Dropdown(options=df['TIME_PERIOD'])

# Display the dropdown widgets
display(countries)
display(time_periods)

def handle_dropdown_change(change):
    selected_country = countries.value
    selected_time_period = time_periods.value

    # Filter the DataFrame based on the selected options
    filtered_df = df[(df['COUNTRY'] == selected_country) & (df['TIME_PERIOD'] == selected_time_period)]

    if filtered_df.empty:
        # Display a message when the filtered DataFrame is empty
        print("The selected country does not have data for the selected time period.")
    else:

        # Create a scatter plot with the filtered data using matplotlib
        plt.scatter(filtered_df['PRICEEACH'], filtered_df['SALES'], c=filtered_df['COUNTRY_CODE'])
        plt.xlabel('Price per Unit')
        plt.ylabel('Sales')
        plt.title('Sales vs. Price per Unit')
        plt.colorbar(label='Country')
        display(filtered_df.describe())


        # Show the plot
        plt.show()

# Register the event handler function to observe changes in the dropdowns
countries.observe(handle_dropdown_change, names='value')
time_periods.observe(handle_dropdown_change, names='value')
```
